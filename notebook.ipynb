<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.1 - Proportion de personnes en sous nutrition</h3>
</div>

```python
# Il faut tout d'abord faire une jointure entre la table population et la table sous nutrition, en ciblant l'année 2017
df_population_2017 = df_population[df_population['Année'] == 2017]
df_sous_nutrition_2017 = df_sous_nutrition[df_sous_nutrition['Année'] == "2016-2018"]
df_merge_popu_sous_nutrition = pd.merge(df_population_2017,df_sous_nutrition_2017, on="Zone", how="inner", suffixes=('population', 'sous_nutrition'))
```

```python
#Affichage du dataset
df_merge_popu_sous_nutrition
```

```python
#Calcul et affichage du nombre de personnes en état de sous nutrition
print(f'Le nombre de personne en sous nutrition est de {int(df_merge_popu_sous_nutrition["sous_nutrition"].sum())}')
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Nombre théorique de personne qui pourrait être nourries</h3>
</div>

```python
#Combien mange en moyenne un être humain ? Source =>
valeur_moyenne_cal = 2300
# Moyenne homme = entre 2400 et 2700
# Moyenne femme = entre 2000 et 2200
```

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
df_dispo_alimentaire_merge = pd.merge(df_dispo_alimentaire, df_population_2017, on="Zone", how='left')
```

```python
#Affichage du nouveau dataframe
df_dispo_alimentaire_merge
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
df_dispo_alimentaire_merge['dispo_kcal'] =  df_dispo_alimentaire_merge['Disponibilité alimentaire (Kcal/personne/jour)'] * df_dispo_alimentaire_merge['Population']
```

```python
#Calcul du nombre d'humains pouvant être nourris
print(df_dispo_alimentaire_merge['dispo_kcal'].sum() * 365)
print(f"le nombre d'humains pouvant être nourris est de {int(df_dispo_alimentaire_merge['dispo_kcal'].sum()  / valeur_moyenne_cal)}")
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.3 - Nombre théorique de personne qui pourrait être nourrie avec les produits végétaux</h3>
</div>

```python
#Transfert des données avec les végétaux dans un nouveau dataframe
df_vegetaux = df_dispo_alimentaire_merge[df_dispo_alimentaire_merge['Origine'] == "vegetale"]
```

```python
#Calcul du nombre de kcal disponible pour les végétaux
print(df_vegetaux['dispo_kcal'].sum() * 365)
print(f"Le nombre de kcal dispo pour les végétaux est de {int(df_vegetaux['dispo_kcal'].sum() / valeur_moyenne_cal )}")
```

```python
#Calcul du nombre d'humains pouvant être nourris avec les végétaux
print(f"le nombre d'humains pouvant être nourris avec les végétaux est de {int(df_vegetaux['dispo_kcal'].sum() / valeur_moyenne_cal)}")
```
